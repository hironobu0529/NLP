```python
# ============================================
# 固定 partition 生成（Part別） for df
#  - Train=80%, Test=20%
#  - ID単位で分割（同一IDは同じpartition）
#  - Age(連続)の分布差 + site/Gender/diagnosisの分布差 + Visit分布差 を最小化
#  - Partごとに partition_P{Part} を列として固定
# ============================================

import pandas as pd
import numpy as np

# ===== パラメータ =====
TEST_SIZE   = 0.20
N_TRIALS    = 2048     # バランス最適化の試行回数（必要なら増やす：1024, 2048など）
GLOBAL_SEED = 20250908

# ===== ユーティリティ =====
def _ensure_cols(df, cols):
    miss = [c for c in cols if c not in df.columns]
    if miss:
        raise ValueError(f"必要な列がありません: {miss}\n現在の列例: {list(df.columns)[:20]} ...")

def _derive_site_from_id(df, id_col, site_col="site"):
    if site_col not in df.columns or df[site_col].isna().all():
        df[site_col] = df[id_col].astype(str).str[1]
    return site_col

def _ks_pvalue_asymptotic(x, y):
    """
    SciPyなしの2標本KSの近似p値（大標本近似）。
    返り値: D統計量, 近似p値
    """
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float)
    x = x[~np.isnan(x)]
    y = y[~np.isnan(y)]
    if len(x) == 0 or len(y) == 0:
        return 0.0, 1.0  # 空なら差なし扱い

    # ECDF差分の最大（D）
    data_all = np.sort(np.concatenate([x, y]))
    cdf_x = np.searchsorted(np.sort(x), data_all, side='right') / len(x)
    cdf_y = np.searchsorted(np.sort(y), data_all, side='right') / len(y)
    D = np.max(np.abs(cdf_x - cdf_y))

    # 近似p値（asymptotic）：p ≈ 2 * sum_{k=1..∞} (-1)^{k-1} exp(-2 k^2 λ^2), λ = D * sqrt(n_eff)
    n_eff = len(x) * len(y) / (len(x) + len(y))
    lam = (np.sqrt(n_eff) + 0.12 + 0.11/np.sqrt(n_eff)) * D  # Stephens correction
    # 打ち切り和
    terms = [(-1)**(k-1) * np.exp(-2 * (k**2) * (lam**2)) for k in range(1, 101)]
    p = 2 * np.sum(terms)
    p = float(np.clip(p, 0.0, 1.0))
    return float(D), p

def _balance_score(df_sub, assign, id_col, cat_cols, age_col, visit_col):
    """
    不均衡スコア（小さいほど良い）
      - Age: 平均差/pooledSD + 分散差/pooledSD + (1 - KS-p)
      - Categorical: ID単位分布差L1 + 行単位分布差L1 の平均を列ごと加算
      - Visit: 行単位の分布差L1
    """
    tmp = df_sub.copy()
    tmp["partition"] = tmp[id_col].map(assign)

    # Age（連続）
    tr = tmp.loc[tmp["partition"]=="Train", age_col].to_numpy()
    te = tmp.loc[tmp["partition"]=="Test",  age_col].to_numpy()
    tr = tr[~np.isnan(tr)]; te = te[~np.isnan(te)]
    if len(tr)==0 or len(te)==0:
        age_term = 1e6
    else:
        mean_diff = abs(np.nanmean(tr) - np.nanmean(te))
        sd_tr = np.nanstd(tr, ddof=1) if len(tr) > 1 else 0.0
        sd_te = np.nanstd(te, ddof=1) if len(te) > 1 else 0.0
        std_diff = abs(sd_tr - sd_te)
        # pooled SD
        if len(tr)+len(te)-2 > 0:
            pooled_var = (((len(tr)-1)*(sd_tr**2) + (len(te)-1)*(sd_te**2)) / (len(tr)+len(te)-2)) if (len(tr)>1 and len(te)>1) else (sd_tr**2 + sd_te**2)/2.0
        else:
            pooled_var = (sd_tr**2 + sd_te**2)/2.0
        pooled_sd = np.sqrt(max(pooled_var, 1e-12))
        # KS近似p
        _, ks_p = _ks_pvalue_asymptotic(tr, te)
        age_term = (mean_diff/pooled_sd) + (std_diff/pooled_sd) + (1.0 - ks_p)

    # カテゴリ
    cat_term = 0.0
    for col in cat_cols:
        # ID単位（被験者カウント）
        idtab = df_sub[[id_col, col]].drop_duplicates(subset=[id_col])
        idtab = idtab.merge(assign.rename("partition"), left_on=id_col, right_index=True, how="left")
        pr_tr = idtab.loc[idtab["partition"]=="Train", col].value_counts(normalize=True)
        pr_te = idtab.loc[idtab["partition"]=="Test",  col].value_counts(normalize=True)
        cats = set(pr_tr.index).union(pr_te.index)
        l1_id = sum(abs(pr_tr.get(c,0)-pr_te.get(c,0)) for c in cats)

        # 行単位（観測数）
        pr_tr_r = tmp.loc[tmp["partition"]=="Train", col].value_counts(normalize=True)
        pr_te_r = tmp.loc[tmp["partition"]=="Test",  col].value_counts(normalize=True)
        cats_r = set(pr_tr_r.index).union(pr_te_r.index)
        l1_row = sum(abs(pr_tr_r.get(c,0)-pr_te_r.get(c,0)) for c in cats_r)

        cat_term += 0.5*l1_id + 0.5*l1_row

    # Visit（行単位）
    visit_term = 0.0
    pv_tr = tmp.loc[tmp["partition"]=="Train", visit_col].value_counts(normalize=True)
    pv_te = tmp.loc[tmp["partition"]=="Test",  visit_col].value_counts(normalize=True)
    cats_v = set(pv_tr.index).union(pv_te.index)
    visit_term = sum(abs(pv_tr.get(c,0)-pv_te.get(c,0)) for c in cats_v)

    # 重み
    return 1.0*age_term + 1.0*cat_term + 0.5*visit_term

def _search_assignment(df_sub, id_col, cat_cols, age_col, visit_col, test_size, n_trials, seed):
    """ランダム割付を多数試行してスコア最小の割付を採用"""
    rng = np.random.default_rng(seed)
    ids = df_sub[id_col].drop_duplicates().tolist()
    n_test = max(1, int(round(len(ids)*test_size)))

    best_score, best_assign = np.inf, None
    for _ in range(n_trials):
        rng.shuffle(ids)
        test_ids = set(ids[:n_test])
        assign = pd.Series({i: ("Test" if i in test_ids else "Train") for i in df_sub[id_col].drop_duplicates()})
        score = _balance_score(df_sub, assign, id_col, cat_cols, age_col, visit_col)
        if score < best_score:
            best_score, best_assign = score, assign
    return best_assign, best_score

def make_partitions_fixed(df,
                          id_col="ID", visit_col="Visit", part_col="Part",
                          age_col="Age", gender_col="Gender", diag_col="diagnosis",
                          site_col="site",
                          test_size=TEST_SIZE, n_trials=N_TRIALS, global_seed=GLOBAL_SEED):
    # 必須列チェック
    _ensure_cols(df, [id_col, visit_col, part_col, age_col, gender_col, diag_col])
    # site（なければIDの2文字目で生成）
    site_col = _derive_site_from_id(df, id_col, site_col)

    # 型整形
    df = df.copy()
    df[site_col]   = df[site_col].astype(str)
    df[gender_col] = df[gender_col].astype(str)
    df[diag_col]   = df[diag_col].astype(str)
    # 🚑 Ageを必ず数値化（文字や欠損はNaNにする）
    df[age_col] = pd.to_numeric(df[age_col], errors="coerce")

    parts = sorted(pd.unique(df[part_col].dropna()))
    out_cols = []

    for p in parts:
        sub = df[df[part_col] == p].copy()
        assign, score = _search_assignment(
            sub,
            id_col=id_col,
            cat_cols=[site_col, gender_col, diag_col],
            age_col=age_col,
            visit_col=visit_col,
            test_size=test_size,
            n_trials=n_trials,
            seed=global_seed + int(p) if str(p).isdigit() else global_seed
        )
        colname = f"partition_P{p}"
        df.loc[df[part_col]==p, colname] = df.loc[df[part_col]==p, id_col].map(assign)
        out_cols.append(colname)

    return df, out_cols


# ===== 実行例（df は既に統合後の DataFrame として存在） =====
# df はそのまま使います
# 例: df = df  # 何もしない。ノートブック上ですでにdfがある前提

# 列名が異なる場合は、以下で指定してください（例：'age' 等）
# df_out, part_cols = make_partitions_fixed(
#     df, id_col="ID", visit_col="Visit", part_col="Part",
#     age_col="Age", gender_col="Gender", diag_col="diagnosis", site_col="site"
# )

# デフォルト列名が合っているならそのまま：
df_out, part_cols = make_partitions_fixed(df)

# 確認＆保存例
print("作成列:", part_cols)
# バランス確認（ダイジェスト）
for c in part_cols:
    print(f"\n=== {c} ===")
    print(df_out[c].value_counts())
    print(df_out.groupby(c)["Age"].agg(['count','mean','std','min','max']).round(3))
    tbl = df_out.pivot_table(index="Visit", columns=c, values="ID", aggfunc="count", fill_value=0)
    print(tbl)

# 固定CSVとして保存（必要なら）
df_out.to_csv("df_with_fixed_partitions.csv", index=False)
```
