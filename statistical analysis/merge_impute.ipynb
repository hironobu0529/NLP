```python
import pandas as pd
import numpy as np

# 一意性チェック
def check_uniqueness(df):
    return df.duplicated(subset=['ID', 'Visit', 'Part']).any()

datasheet_A = pd.read_csv("Dataframe_sBERT_R6Nov22.csv")
datasheet_B = pd.read_csv("Dataframe_w2v_adj_R6Dec5.csv")
datasheet_C = pd.read_csv("Dataframe_AnalysisGINZA_R7Aug16.csv")
datasheet_D = pd.read_csv("w2Vsimilarity_r&b_R6Nov29.csv")
datasheet_E1 = pd.read_csv("wvGraphAnalyzer_part1_R7Mar13_N10_sth0.3.csv")
datasheet_E2 = pd.read_csv("wvGraphAnalyzer_part2_R7Mar13_N10_sth0.3.csv")
datasheet_E3 = pd.read_csv("wvGraphAnalyzer_part3_R7Mar14_N10_sth0.3.csv")
datasheet_F = pd.read_csv("Dataframe_sentenceSimilarity_r&b_R7Feb17.csv")
datasheet_G = pd.read_csv("Dataframe_analyzeNestedDependency_integration_R7Aug12_RecordVoice_vf_CEJCminus.csv")
```

```python
datasheet_E = pd.concat([datasheet_E1, datasheet_E2, datasheet_E3], ignore_index=True)
```

```python
# 各データシートのデータ数を表示
print("Datasheet A size:", datasheet_A.shape)
print("Datasheet B size:", datasheet_B.shape)
print("Datasheet C size:", datasheet_C.shape)
print("Datasheet D size:", datasheet_D.shape)
print("Datasheet E size:", datasheet_E.shape)
print("Datasheet F size:", datasheet_F.shape)
print("Datasheet G size:", datasheet_G.shape)


# 一意性を確認
print("Uniqueness in A:", not check_uniqueness(datasheet_A))
print("Uniqueness in B:", not check_uniqueness(datasheet_B))
print("Uniqueness in C:", not check_uniqueness(datasheet_C))
print("Uniqueness in D:", not check_uniqueness(datasheet_D))
print("Uniqueness in E1:", not check_uniqueness(datasheet_E1))
print("Uniqueness in E2:", not check_uniqueness(datasheet_E2))
print("Uniqueness in E3:", not check_uniqueness(datasheet_E3))
print("Uniqueness in E:", not check_uniqueness(datasheet_E))
print("Uniqueness in F:", not check_uniqueness(datasheet_F))
print("Uniqueness in G:", not check_uniqueness(datasheet_G))
```

```python
# 重複列削除関数
def remove_duplicate_columns(df):
    columns_to_drop = [col for col in df.columns if 'Unnamed' in col]
    if columns_to_drop:
        df = df.drop(columns=columns_to_drop)
    return df


datasheet_A = remove_duplicate_columns(datasheet_A)
datasheet_B = remove_duplicate_columns(datasheet_B)
datasheet_C = remove_duplicate_columns(datasheet_C)
datasheet_D = remove_duplicate_columns(datasheet_D)
datasheet_E = remove_duplicate_columns(datasheet_E)
datasheet_F = remove_duplicate_columns(datasheet_F)
datasheet_G = remove_duplicate_columns(datasheet_G)
```

```python
# 各データシートのデータ数を表示
print("Datasheet A size:", datasheet_A.shape)
print("Datasheet B size:", datasheet_B.shape)
print("Datasheet C size:", datasheet_C.shape)
print("Datasheet D size:", datasheet_D.shape)
print("Datasheet E size:", datasheet_E.shape)
print("Datasheet F size:", datasheet_F.shape)
print("Datasheet G size:", datasheet_G.shape)


# 一意性を確認
print("Uniqueness in A:", not check_uniqueness(datasheet_A))
print("Uniqueness in B:", not check_uniqueness(datasheet_B))
print("Uniqueness in C:", not check_uniqueness(datasheet_C))
print("Uniqueness in D:", not check_uniqueness(datasheet_D))
print("Uniqueness in E1:", not check_uniqueness(datasheet_E1))
print("Uniqueness in E2:", not check_uniqueness(datasheet_E2))
print("Uniqueness in E3:", not check_uniqueness(datasheet_E3))
print("Uniqueness in E:", not check_uniqueness(datasheet_E))
print("Uniqueness in F:", not check_uniqueness(datasheet_F))
print("Uniqueness in G:", not check_uniqueness(datasheet_G))
```

```python
# 共通のキーでデータを外部結合（マージ）
merged_df = datasheet_A.merge(datasheet_B, on=['ID', 'Visit', 'Part'], how='outer', validate='one_to_one')
merged_df = merged_df.merge(datasheet_C, on=['ID', 'Visit', 'Part'], how='outer', validate='one_to_one')
merged_df = merged_df.merge(datasheet_D, on=['ID', 'Visit', 'Part'], how='outer', validate='one_to_one')
merged_df = merged_df.merge(datasheet_E, on=['ID', 'Visit', 'Part'], how='outer', validate='one_to_one')
merged_df = merged_df.merge(datasheet_F, on=['ID', 'Visit', 'Part'], how='outer', validate='one_to_one')
merged_df = merged_df.merge(datasheet_G, on=['ID', 'Visit', 'Part'], how='outer', validate='one_to_one')
```

```python
# 結合後のデータ数を表示
print("Merged DataFrame size:", merged_df.shape)

print("Uniqueness in merge:", not check_uniqueness(merged_df))
```
